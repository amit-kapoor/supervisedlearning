```python
import pandas as pd
import numpy as np
```

```python
# read the automobile data
auto_data = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data", header=None)
auto_data.head()
```

```python
auto_data.columns
```

```python
# create headers list and insert in dataframe
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]

auto_data.columns = headers
auto_data.head()
```

```python
auto_data = auto_data.replace('?', np.nan)
auto_data.head()
```

```python
auto_data.describe()
```

```python
auto_data.describe(include='all')
```

```python
auto_data['price'].describe()
```

```python
auto_data.dtypes
```

```python
# convert price column to numeric
auto_data['price'] = pd.to_numeric(auto_data['price'], errors='coerce')
auto_data['price'].describe()
```

```python
# convert horsepower column to numeric
auto_data['horsepower'] = pd.to_numeric(auto_data['horsepower'], errors='coerce')
auto_data['horsepower'].describe()
```

```python
# convert normalized-losses column to numeric
auto_data = auto_data.drop('normalized-losses', axis=1)
auto_data.head()
```

```python
# print unique values of num-of-cylinders
np.unique(auto_data['num-of-cylinders'])
```

```python
# replace num-of-cylinders values to corressponding numbers
cylinders_dict = {'eight': 8,
                  'five':5,
                  'four':4,
                  'six':6,
                  'three':3,
                  'twelve':12,
                  'two':2}
auto_data['num-of-cylinders'].replace(cylinders_dict, inplace=True)
auto_data.head()
```

```python
# One hot encoding
auto_data = pd.get_dummies(auto_data, 
                           columns=['make',
                                    'fuel-type',
                                    'aspiration',
                                    'num-of-doors',
                                    'body-style',
                                    'drive-wheels',
                                    'engine-location',
                                    'engine-type',
                                    'fuel-system'])

auto_data.head()
```

```python
# Drop all rows having NaN values
auto_data = auto_data.dropna()
auto_data
```

```python
# check for any null values
auto_data[auto_data.isnull().any(axis=1)]
```

```python
# split training and test data
from sklearn.model_selection import train_test_split

X = auto_data.drop('price', axis=1)
y = auto_data['price']

# splitting into 75% for training and 25% for testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=41)
```

```python
from sklearn.linear_model import LinearRegression

linear_regression = LinearRegression()
linear_regression.fit(X_train, y_train)
```

```python
# R-Square
linear_regression.score(X_train, y_train)
```

```python
linear_regression.coef_
```

```python
predictors = X_train.columns

weights_with_cols = pd.Series(linear_regression.coef_, predictors).sort_values()
weights_with_cols
```

## Coefficients with -ve values show corressponding features are negatively correlated and cars with those features are cheaper. On the other side cars with features like make_bmw, make_porche etc are expensive

```python
y_predict = linear_regression.predict(X_test)
```

```python
# Draw the graph for actual and predicted values

import matplotlib.pyplot as plt
import seaborn as sns

plt.plot(y_predict, label='Predicted')
plt.plot(y_test.values, label='Actual')

plt.ylabel('Price')
plt.legend()

plt.show()
```

```python
# R-square for test 
r_2 = linear_regression.score(X_test, y_test)
r_2
```

```python
from sklearn.metrics import mean_squared_error

linear_regression_mse = mean_squared_error(y_predict, y_test)
linear_regression_mse
```

```python
# RMSE - root mean square error

import math

math.sqrt(linear_regression_mse)
```

## Lasso Regression

```python
from sklearn.linear_model import Lasso

lasso_reg = Lasso(alpha=5, normalize=True)
lasso_reg.fit(X_train, y_train)
```

```python
lasso_reg.score(X_train, y_train)
```

```python
lasso_coeff = pd.Series(lasso_reg.coef_, predictors).sort_values()
print(lasso_coeff)
```

```python
y_predict = lasso_reg.predict(X_test)
```

```python
from matplotlib import pylab

pylab.rcParams['figure.figsize']=(15,6)

plt.plot(y_predict, label='Predicted')
plt.plot(y_test.values, label='Actual')

plt.ylabel('Price')
plt.legend()
plt.show()
```

```python
r_square = lasso_reg.score(X_test, y_test)
r_square
```

```python
lasso_reg_mse = mean_squared_error(y_predict, y_test)
lasso_reg_mse
```

```python
math.sqrt(lasso_reg_mse)
```
